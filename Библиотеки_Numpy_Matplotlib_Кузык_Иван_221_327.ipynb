```python
def multiply_matrices(matrix1, matrix2, matrix3):
  """
  Функция умножения трех матриц без использования встроенных средств np.dot() и @.

  Args:
    matrix1: Первая матрица (список списков).
    matrix2: Вторая матрица (список списков).
    matrix3: Третья матрица (список списков).

  Returns:
    Результирующая матрица (список списков) или None, если умножение невозможно.
  """

  # Проверка возможности умножения matrix1 и matrix2
  if len(matrix1[0]) != len(matrix2):
    return None

  # Проверка возможности умножения результата (matrix1 * matrix2) и matrix3
  if len(matrix2[0]) != len(matrix3):
    return None

  # Умножение matrix1 и matrix2
  result12 = [[0 for _ in range(len(matrix2[0]))] for _ in range(len(matrix1))]
  for i in range(len(matrix1)):
    for j in range(len(matrix2[0])):
      for k in range(len(matrix2)):
        result12[i][j] += matrix1[i][k] * matrix2[k][j]

  # Умножение результата (result12) и matrix3
  result = [[0 for _ in range(len(matrix3[0]))] for _ in range(len(result12))]
  for i in range(len(result12)):
    for j in range(len(matrix3[0])):
      for k in range(len(matrix3)):
        result[i][j] += result12[i][k] * matrix3[k][j]

  return result

# Пример использования:
matrix1 = [[1, 2], [3, 4]]
matrix2 = [[5, 6], [7, 8]]
matrix3 = [[9, 10], [11, 12]]

result = multiply_matrices(matrix1, matrix2, matrix3)

if result:
  for row in result:
    print(row)
else:
  print("Умножение невозможно")
```

```python
def spiral_matrix(n):
  """
  Функция выводит на печать матрицу размером N*N, заполненную числами
  натурального ряда, начиная с первого элемента и двигаясь по спирали.

  Args:
    n: Размер матрицы (целое число).
  """

  matrix = [[0 for _ in range(n)] for _ in range(n)]
  num = 1
  top, bottom, left, right = 0, n - 1, 0, n - 1

  while num <= n * n:
    # Движение вправо
    for i in range(left, right + 1):
      matrix[top][i] = num
      num += 1
    top += 1

    # Движение вниз
    for i in range(top, bottom + 1):
      matrix[i][right] = num
      num += 1
    right -= 1

    # Движение влево
    if top <= bottom:
      for i in range(right, left - 1, -1):
        matrix[bottom][i] = num
        num += 1
      bottom -= 1

    # Движение вверх
    if left <= right:
      for i in range(bottom, top - 1, -1):
        matrix[i][left] = num
        num += 1
      left += 1

  # Вывод матрицы
  for row in matrix:
    print(row)

# Пример использования:
n = int(input("Введите размер матрицы (N): "))
spiral_matrix(n)
```

```python
import numpy as np

def weekdays_in_february(year):
  """
  Функция находит количество будних дней в феврале заданного года,
  используя библиотеку NumPy.

  Args:
    year: Год (целое число).

  Returns:
    Количество будних дней в феврале.
  """

  start_date = np.datetime64(f'{year}-02-01')
  end_date = np.datetime64(f'{year}-03-01')
  days = np.arange(start_date, end_date, dtype='datetime64[D]')
  weekdays = np.is_busday(days)
  return np.count_nonzero(weekdays)

# Пример использования:
year = 2015
weekdays_count = weekdays_in_february(year)
print(f"Количество будних дней в феврале {year} года: {weekdays_count}")
```
