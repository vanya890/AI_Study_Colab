Задание 1. Создать формальную модель нейрона

Нейрон МакКалока-Питтса

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
def McCuloh_Pitts_neuron(x = None, w = None, bias = 0):
  # w - веса
  # x - входы
  #print('X: ',x)
  #print('W: ',w)
  #print('bias: ',bias)
  s = np.dot(w, x.T )+ bias
  #print('суммарный уровень входов нейрона на их веса: ',s)
  y = np.sign(s)
  plt.plot(y)
  return y
```

Запустим нейрон на примере х с весами w

```python
x = np.array([1,0])
w = np.array([1,-1])
y_out = McCuloh_Pitts_neuron(x = x, w = w, bias = 0.3)
print('ответ нейрона',y_out)
```

Линейный нейрон

```python
def linear_neuron(x = None, w = None, bias = 0):
  #print('X: ',x)
  #print('W: ',w)
  #print('bias: ',bias)
  s = np.dot(w, x.T )+ bias
  #print('суммарный уровень входов нейрона на их веса: ',s)
  y = s
  #print("Производная: ", diff(z,z))
  plt.plot(y)
  return y
```

Запустим нейрон на множестве примеров в х

```python
x = np.random.random((10,2))
w = np.array([1,-1])
y_out1 = linear_neuron(x = x, w = w, bias = 0.3)
print('ответ нейрона',y_out1)

plt.show()

xp = np.linspace(-10, 10)  # 100 точек от -5 до 5
yp = xp
plt.plot(xp, yp)
plt.title("Производная функции")
plt.show()
```

Сигмоидный нейрон

```python
def sigmoid_neuron(x = None, w = None, bias=0):
  #print('X: ',x)
  #print('W: ',w)
  #print('bias: ',bias)
  s = np.dot(w, x.T )+ bias
  #print('суммарный уровень входов нейрона на их веса: ',s)
  y = 1 / (1 + np.exp(-s))
  #print("Производная: ",diff((1 / (1 + exp(-z))),z))
  plt.plot(y, 'r')
  return y
```

Тангенциальный

```python
def sigmoidt_neuron(x = None, w = None, bias=0):
  #print('X: ',x)
  #print('W: ',w)
  #print('bias: ',bias)
  s = np.dot(w, x.T )+ bias
  #print('суммарный уровень входов нейрона на их веса: ',s)
  y = 2 / (1 + np.exp(-s))-1
  plt.plot(y)
  return y
```

Запустим нейрон

```python
y_out2 = sigmoid_neuron(x = x, w = w, bias = 0.3)
print('ответ нейрона',y_out2)
plt.show()

xp = np.linspace(-10, 10)  # 100 точек от -5 до 5
yp = np.exp(-xp)/(1 + np.exp(-xp))**2
plt.plot(xp, yp)
plt.title("Производная функции")
plt.show()
```
